```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Janvar","Fevral","Mart","Aprel","Afgust"]
y= [10,11,12,14,15]
plt.figure(figsize=(6,5))
plt.bar(x,y, label = "Стобчатая диаграмма", color = "orange")
plt.plot(x,y, label = "Линейный график")
plt.legend()
plt.title("Комбинированный график")
plt.xlabel("Ось x")
plt.ylabel("Ось Y")

plt.show()
```

```python
x = ["Janvar","Fevral","Mart","Aprel","Afgust"]
y= [10,11,12,14,19]
colors = ["purple", "green","pink", "blue", "yellow"]
plt.pie(y, labels=x, colors= colors, autopct= "%1.1f%%")
plt.title("Круговая диаграмма")
plt.show()

```
